# color mask

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = './img/cube.png'

img = cv2.imread(path)
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
hsv = cv2.cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

# red 마스크
lowred = np.array([0,50,50])
upred = np.array([10,255,255])
mask = cv2.inRange(hsv,lowred,upred)
red_ms = cv2.bitwise_and(rgb,rgb.copy(),mask=mask)

# orange 마스크
loworan = np.array([11,50,50])
uporan = np.array([20,255,255])
mask2 = cv2.inRange(hsv,loworan,uporan)
oran_ms = cv2.bitwise_and(rgb,rgb.copy(),mask=mask2)


imgs = {'rgb':rgb, 'hsv':hsv, 'red':red_ms, 'orange':oran_ms}
for i, (key,value) in enumerate(imgs.items()):
    plt.subplot(2,2,i+1)
    plt.title(key)
    plt.imshow(value)
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```
